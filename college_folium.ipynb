##Import Dataset

```python
dataset = pd.read_csv('college.csv')
print(dataset)
```

## Marking the IITs on Map

```python
map_college = fl.Map(
    location = [20.5937, 78.9629],
    zoom_start = 4.5,
    tiles = 'OpenStreetMap')

college = dataset.iloc[0]
for _, clg in dataset.iterrows():
  fl.Marker(
      location = [clg['Longitude'], clg['Latitude']],
      tooltip = clg['Name']
  ).add_to(map_college)

map_college
```
